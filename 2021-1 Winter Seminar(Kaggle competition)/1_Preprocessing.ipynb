# 1. Preprocessing

데이터 로드

```python
train = pd.read_csv('train.csv')
```

```python
pd.set_option('display.max_columns',None) # ...으로 요약 안 되게
```

```python
train_y = train['class']
train_x = train.drop(['class'], axis=1)
```

column, row NA 비율 구하기

```python
na_ratio_col = train_x.isnull().sum().sort_values(ascending=False)/len(train_x)
na_ratio_row = train_x.isnull().sum(1).sort_values(ascending=False)/len(train_x.columns)
```

### 1.1 column, row 둘 다 NA 비율 50% 이상인 항목 삭제

### train

* column

```python
col_del = [ na_ratio_col.index[i]  for i in range(0,len(na_ratio_col)) if ( na_ratio_col[i] > 0.5 ) == True]
train_x.drop(col_del, axis=1, inplace=True)
len(train_x.columns)
```

* row - x (features)

```python
row_del = [ na_ratio_row.index[i] for i in range(0,len(na_ratio_row)) if (na_ratio_row[i] <= 0.5 ) ]
train_x = train_x.iloc[row_del,:]
len(train_x)
```

* row - y (타겟 변수)

```python
train_y = train_y.iloc[row_del]
len(train_y)
```

### 1.2 Scaling

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
train_x_scaled = pd.DataFrame(scaler.fit_transform(train_x), columns = train_x.columns)
```

### 1.3 KNN Imputation

```python
from sklearn.impute import KNNImputer
```

```python
knn = KNNImputer(n_neighbors=5)
train_x_knn = knn.fit_transform(train_x_scaled)
```

```python
knn_df = pd.DataFrame(train_x_knn, columns=train_x_scaled.columns)
knn_df
```

```python
#knn_df.to_csv('knn_imputation_final.csv')
```

### 1.4 SMOTE
불균형 처리

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE(random_state=500)
```

```python
knn_nopca_x_smote, knn_nopca_y_smote = smote.fit_sample(knn_df, train_y)
```

### 1.5 저장

```python
knn_nopca_x_smote_df = pd.DataFrame(knn_nopca_x_smote, columns =knn_df.columns )
```

```python
knn_nopca_y_smote_df = pd.DataFrame(knn_nopca_y_smote, columns =['class'])
```

```python
knn_nopca_x_smote_df.to_csv('knn_nopca_x_smote_final2.csv')
knn_nopca_y_smote_df.to_csv('knn_y_smote_final2.csv')
```
